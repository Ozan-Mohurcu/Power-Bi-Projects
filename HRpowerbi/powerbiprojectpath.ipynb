# 📊 İnsan Kaynakları Analitik Veri Seti  

Bu veri seti, bir İK (İnsan Kaynakları) analitik çalışmasına aittir ve her bir çalışanın çeşitli özelliklerini içeren **38 sütundan** oluşmaktadır.  

## 🏢 Sütunlar ve Açıklamaları  

- **🆔 EmpID:** Çalışanın benzersiz kimlik numarası.  
- **🎂 Age:** Çalışanın yaşı.  
- **📊 AgeGroup:** Çalışanın yaş aralığı (örn: 18-25).  
- **🔄 Attrition:** İşten ayrılma durumu ("Yes" / "No").  
- **🛄 BusinessTravel:** İş seyahati sıklığı (🚶 Non-Travel, ✈️ Travel_Rarely, 🌍 Travel_Frequently).  
- **💰 DailyRate:** Çalışanın günlük kazancı.  
- **🏢 Department:** Çalıştığı departman (Sales, R&D, vb.).  
- **📍 DistanceFromHome:** Ev-iş arası mesafe (km).  
- **🎓 Education:** Eğitim seviyesi (1 = Ortaokul, 5 = Doktora).  
- **📘 EducationField:** Eğitim alanı (Life Sciences, Medical, vb.).  
- **👥 EmployeeCount:** Çalışan sayısı (muhtemelen hep **1**).  
- **🔢 EmployeeNumber:** Çalışanın benzersiz numarası.  
- **🏡 EnvironmentSatisfaction:** İş ortamı memnuniyeti (1-4).  
- **🚻 Gender:** Cinsiyet (🚹 Male / 🚺 Female).  
- **⏳ HourlyRate:** Saatlik ücret.  
- **🔗 JobInvolvement:** İşe katılım seviyesi (1-4).  
- **📈 JobLevel:** İş pozisyon seviyesi.  
- **💼 JobRole:** Çalışanın iş rolü (Manager, Sales Rep, vb.).  
- **😊 JobSatisfaction:** İş memnuniyeti (1-4).  
- **💍 MaritalStatus:** Medeni durum (❤️ Single, 💑 Married, 💔 Divorced).  
- **💵 MonthlyIncome:** Aylık gelir.  
- **📊 SalarySlab:** Maaş kategorisi.  
- **💳 MonthlyRate:** Aylık maaş.  
- **🏢 NumCompaniesWorked:** Önceki işyeri sayısı.  
- **🆗 Over18:** 18 yaş üstü olup olmadığı (Muhtemelen hep "Yes").  
- **⏰ OverTime:** Fazla mesai yapma durumu (Yes / No).  
- **📈 PercentSalaryHike:** Maaş artış yüzdesi.  
- **⭐ PerformanceRating:** Performans değerlendirmesi (1-4).  
- **🤝 RelationshipSatisfaction:** İş yeri ilişkileri memnuniyeti (1-4).  
- **⌛ StandardHours:** Standart çalışma süresi (Muhtemelen hep **80**).  
- **📈 StockOptionLevel:** Hisse opsiyonu seviyesi.  
- **👴 TotalWorkingYears:** Toplam çalışma yılı.  
- **📚 TrainingTimesLastYear:** Son yıl alınan eğitim sayısı.  
- **⚖️ WorkLifeBalance:** İş-yaşam dengesi (1-4).  
- **🏢 YearsAtCompany:** Şirkette geçirilen yıl sayısı.  
- **🎭 YearsInCurrentRole:** Mevcut rolde geçirilen yıl sayısı.  
- **🎖 YearsSinceLastPromotion:** Son terfiden bu yana geçen yıl.  
- **👨‍💼 YearsWithCurrManager:** Mevcut yöneticiyle çalışma süresi (yıl).  

# 📊 Power BI İnsan Kaynakları Analitik Dashboard Projesi

## 1️⃣ Proje Planlama ve Hazırlık  

### 1.1 Veri Hazırlığı  
- 📥 **Veri Yükleme:** Veri setini Power BI'a aktarın.  
- 🛠 **Veri Temizliği ve Dönüştürme:**  
  - Gereksiz sütunları kaldırın (örn: `EmployeeCount`, `Over18`, `StandardHours`).

Dashboard 3-4 sayfa içerecektir:

2. **Maaş ve Finansal Analiz**

### 🎛 **Filtreler:**

## 3️⃣ Sayfa 2: Maaş ve Finansal Analiz

# 4. Sayfa: İş Seyahatleri ve Uzaklık Analizi
**Amaç:** Çalışanların iş seyahatleri sıklığı ve evden işe uzaklıklarının analizi.

## Yapılacak Görselleştirmeler:

### **BusinessTravel (İş Seyahati):**
- İş seyahat sıklıklarının dağılımını bir **pasta grafik** ile gösterin.
- Seyahat sıklığına göre **memnuniyet** veya **işten ayrılma oranlarını (Attrition)** bir **çubuk grafik** ile inceleyin.

### **DistanceFromHome (Evden İşe Uzaklık):**
- Evden işe olan uzaklık ile işten ayrılma arasındaki ilişkiyi bir **dağılım grafiği (scatter plot)** ile analiz edin.
- Uzaklık aralıklarını bir **histogram** ile görselleştirin.

---

# 5. Sayfa: Performans ve Kariyer Gelişimi
**Amaç:** Çalışanların performansı, eğitim alma sıklıkları ve kariyer gelişim süreçlerini analiz edin.

## Yapılacak Görselleştirmeler:

### **PerformanceRating (Performans Değerlendirme):**
- Performans derecelendirmelerinin dağılımını bir **sütun grafik** ile gösterin.

### **TrainingTimesLastYear (Geçen Yıl Alınan Eğitimler):**
- Eğitim sıklığı ile **performans** veya **memnuniyet** arasındaki ilişkiyi inceleyin.

### **YearsSinceLastPromotion (Son Terfiden Bu Yana Geçen Yıl):**
- Son terfi süresi ile **işten ayrılma oranlarını** bir **çubuk grafik** ile gösterin.

### **JobLevel (İş Seviyesi):**
- İş seviyesi ile **toplam çalışma yılları** arasındaki ilişkiyi analiz edin.

---

# 6. Sayfa: Fazla Mesai (Overtime) ve İş-Yaşam Dengesi
**Amaç:** Fazla mesainin çalışanların memnuniyeti, maaşı ve işten ayrılma oranı üzerindeki etkisini analiz edin.

## Yapılacak Görselleştirmeler:

### **Overtime (Fazla Mesai):**
- Fazla mesai yapan ve yapmayan çalışanların oranını bir **pasta grafik** ile gösterin.
- Fazla mesai yapanların **maaş ortalamasını** bir **KPI veya çubuk grafik** ile sunun.

### **WorkLifeBalance (İş-Yaşam Dengesi):**
- İş-yaşam dengesi puanlarının dağılımını bir **çubuk grafik** ile görselleştirin.
- Fazla mesai yapanların **iş-yaşam dengesi puanlarını** karşılaştırın.

---

# 7. Sayfa: Çalışanların İşe Bağlılık Süreçleri
**Amaç:** Çalışanların şirkette kalma süreleri, rolleri ve yöneticileriyle ilişkilerini analiz edin.

## Yapılacak Görselleştirmeler:

### **YearsAtCompany (Şirkette Geçirilen Yıllar):**
- Şirkette geçirilen yıllara göre çalışan dağılımını **histogram** ile analiz edin.

### **YearsWithCurrManager (Mevcut Yönetici ile Çalışma Süresi):**
- Yönetici ile çalışma süresi ile **işten ayrılma oranı** arasındaki ilişkiyi analiz edin.

### **JobRole ve YearsInCurrentRole:**
- Mevcut iş rolünde geçirilen yılları görselleştirin.
- Rol ve şirket içindeki bağlılık süresini karşılaştırmak için bir **yığınlanmış grafik** kullanın.

---

# 8. Sayfa: Çalışanların Maaş Artışı ve Performans Analizi

## Yapılacak Görselleştirmeler:

### **PercentSalaryHike (Maaş Zammı Oranı):**
- Maaş artış oranlarını bir **histogram** veya **çubuk grafik** ile görselleştirin.

### **PerformanceRating ve PercentSalaryHike:**

### **MonthlyIncome ve PercentSalaryHike:**
